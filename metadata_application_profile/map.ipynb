```python
import requests
```

```python
lh_endpoint = 'https://www.loc.gov/free-to-use/lighthouses/'

parameters = {
    'fo' :'json',
    'fa' : 'online-format:image'
}
```

```python
r = requests.get(lh_endpoint, params=parameters)
```

```python
data = r.json()
print(data)
```

```python
for key in data.keys():
    print(key)
    
```

```python
data['content']['set']['items']
```

```python
link_list = []

for result in data['content']['set']['items']:
    link = ('https://loc.gov' + result['link'])
    link_list.append(link)
print(link_list)
```

```python
for link in link_list:
    r_lh = requests.get(link, params=parameters)
    lh_data = r_lh.json()
    print(lh_data)
```
